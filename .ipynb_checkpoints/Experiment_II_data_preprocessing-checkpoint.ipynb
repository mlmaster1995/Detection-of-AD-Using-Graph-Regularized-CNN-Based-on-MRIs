### Define Path, Rename Images and Count Images for Both Train and Test

```python
''' label definition: NonDemented - 0, VeryMildDemented - 1, MildDemented -2, ModerateDemented -3 '''
label_list = ['NonDemented', 'VeryMildDemented', 'MildDemented', 'ModerateDemented']
root_path = './project_dataset/graph_images/'
train_root_path = f'{root_path}train/'
test_root_path = f'{root_path}test/'

train_path_list = [f'{train_root_path}{label}/' for label in label_list]
train_tfr_list = [f'{train_root_path}{label}.tfr' for label in label_list]

test_path_list = [f'{test_root_path}{label}/' for label in label_list]
test_tfr_list = [f'{test_root_path}{label}.tfr' for label in label_list]
```

```python
'''rename images in the graph_image folder'''
# GraphDataProcess.rename_images(label_list, train_root_path) 
# GraphDataProcess.rename_images(label_list, test_root_path) 

'''count each label images'''
NUM_train=[]
NUM_test=[]
for label in label_list: NUM_train.append(len(os.listdir(f'{train_root_path}{label}/')))
for label in label_list: NUM_test.append(len(os.listdir(f'{test_root_path}{label}/')))
print("train image count:", NUM_train, "\ntest image count:", NUM_test)
```

```python
'''test data: sample image plotting'''
plt.figure(figsize=(20,20))
for index, label in enumerate(label_list):
    image_string = open(f'{test_root_path}{label_list[index]}/{index}_0.jpg', 'rb').read()
    image_tensor = tf.io.decode_jpeg(contents=image_string, channels=3, name=None)
    image_tensor = tf.image.resize(image_tensor, size=(100,100))
    ax = plt.subplot(1, 4, index + 1)
    plt.imshow(image_tensor/255)
    plt.title(label_list[index])
```

### Balance Train Dataset

```python
'''image counts before balancing'''
image_counts={}
for label in label_list: 
    image_counts[f'{label}_image_counts'] = len(os.listdir(f'{train_root_path}{label}/'))
count_list = list(image_counts.values())
plt.figure(figsize=(10,5))
plt.bar(label_list, count_list)
plt.grid()
plt.show()
```
